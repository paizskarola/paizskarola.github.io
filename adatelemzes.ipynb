```
duplicated_rows = df[df.duplicated()]
num_duplicated_rows = duplicated_rows.shape[0]
num_duplicated_rows

df.drop_duplicates(inplace=True)

df.shape
```

```
year_counts =  df['work_year'].value_counts()
year_counts
```

```
df = df[ df['work_year'] > 2020]
```

```
df['remote_ratio'] = df['remote_ratio'].replace({0: 'on site', 50: 'hybrid', 100: 'remote'})
```

```
df['job_title'] = df['job_title'].replace({'Machine Learning Engineer':'ML Engineer'})
```

```
companies_loc =  df['company_location'].value_counts()
```

```
employees_contries =  df['employee_residence'].value_counts()
```

```
df['company_location'] = df['company_location'].replace({'IL':'PS'})
df['employee_residence'] = df['employee_residence'].replace({'IL':'PS'})
```

```
df.to_csv('adatforras_tisztitott.csv')
files.download("adatforras_tisztitott.csv")
```
